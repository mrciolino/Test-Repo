```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
import pyLDAvis.sklearn
import pandas as pd
import nltk

# settings
pyLDAvis.enable_notebook()

# read in dataframe
df = pd.read_csv("../../Data/PreProcessed/2017-2018/bills_preprocessed.csv")

# reduce data for tests
df = df[0:2000]
```

```python
# preprocessing step to reduce words to semi-root form


# build ngram-tf (no stop words, no accents, no numbers, 2 word combos, lowercase)
vectorizer = CountVectorizer(ngram_range=(1, 2),
                             lowercase=True,
                             stop_words='english',
                             analyzer='word',
                             strip_accents="ascii",
                             token_pattern=r'[a-zA-Z\-][a-zA-Z\-]{2,}')

# Build a Latent Dirichlet Allocation Model
lda_model = LatentDirichletAllocation(n_components=25,
                                      max_iter=25,
                                      learning_method='online')
```

```python
def topic_model(lda_model, vectorizer, data):
    
    # count 2-gram occures of words into matrix
    vectorized = vectorizer.fit_transform(data)
    
    # use lda to transoform into clusters
    lda_model.fit_transform(vectorized)
    
    # visulzie vlusting with pyLDAvis
    panel = pyLDAvis.sklearn.prepare(lda_model, vectorized, vectorizer, mds='tsne')
    
    return panel
```

```python
# Title Viz
topic_model(lda_model, vectorizer, df.title.values.astype('U'))
```
